## Principal Component Analysis

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
plt.rcParams['figure.figsize'] = [6,6]
```

```python
# Reading the dataset
data = pd.read_csv('wdbc.csv',header=None)
# Class 1 = Maligant Tumor
# Class 2 = Benign Tumor
data.loc[data[1] == "M",1] = 1
data.loc[data[1] == "B",1] = 0
x = data.to_numpy()
X = x[:,2:]
y = x[:,1]

# Number of datapoints
N = len(X)
# Number of Features
p = len(X[0])
```

```python
# Calculating mean of data
mean = np.asmatrix(np.mean(X,axis = 0)).T
# Now making data ZERO-MEAN (i.e Standardized data) as required for Principle Component Analysis
Xnew = X - mean.T
```

```python
# Covariance matrix
Cxx= (1/N)*(np.dot(Xnew.T,Xnew))
print("The Covariance matrix of Given data is")
Cxx = np.array(Cxx, dtype = float)
```

```python
D,E = np.linalg.eig(Cxx)
print("Y=PX, where Y is representation of data X (in reduced dimension). By eigen value decomposition",end="")
print(", we can find the transformation matrix P. P=E.T (Transpose) E=Eigen Vector Matrix.")
#print("Eigen Vector matrix(E) of Cxx is")
#print(E)
print("The Eigen Values of matrix Cxx is")
print(D)
```

```python
P = E
Y = np.dot(Xnew , P)
print(Y)
print("\n\nFrom above Transformed data Y we can see that for columns 1 to 7, values are large and should not be neglected.")
print("However rest all columns have very small values and can be approximated to 0")
print("In this way, we have represented the data having 30 features to data having 7 features")
```

```python
# Plotting the results
Ynew = Y
Ynew = np.hstack((Ynew,np.asmatrix(y).T))

condlist = [Ynew[:,30]==1]
choicelist = [Ynew]
Y1 = np.select(condlist, choicelist)
Y1 = Y1[~np.all(Y1 == 0, axis=1)]

condlist = [Ynew[:,30]==0]
choicelist = [Ynew]
Y2 = np.select(condlist, choicelist)
Y2 = Y2[~np.all(Y2 == 0, axis=1)]

plt.scatter([Y1[:,0]],[Y1[:,1]],color="red",edgecolor="black",label="Maligant Tumor")
plt.scatter([Y2[:,0]],[Y2[:,1]],color="blue",edgecolor="black",label="Benign Tumor")
plt.title("Principle Component Analysis on Breast Cancer Wisconsin Dataset")
plt.xlabel("Principal Component 1")
plt.ylabel("Principal Component 2")
plt.legend()
plt.show()
```

```python
# Selecting 7 Features
Xnew = Y[:,:7]
Xnew = np.asarray(Xnew)
Xnew = Xnew.astype("float64")
y = y.reshape(Xnew.shape[0],1)
y = y.astype("float64")
data = np.hstack((Xnew, y))

train = data[:455]
test = data[455:]

N,p = train.shape
Nt,pt = test.shape

print("Train Set:",N,p)
print("Test Set:",Nt,pt)

train_X = train[:,:-1]
train_y = train[:,-1]
train_y = train_y.reshape(N,1)

test_X = test[:,:-1]
test_y = test[:,-1]
test_y = test_y.reshape(Nt,1)

train_X = np.hstack((np.ones((N,1)),train_X))
test_X = np.hstack((np.ones((Nt,1)),test_X))
```

```python
def sigmoid(a):
    return(1/(1+np.exp(-a)))
```

```python
# Logistic Regression Gradient Descent
# Gradient Descent
iterations = 500
alpha = 0.00000001

theta = (0.001)*np.random.rand(p)
theta = np.asmatrix(theta).T

h_theta = sigmoid(np.dot(train_X,theta))
cost = []
#iterat = []

for i in range(iterations):
    h_theta = sigmoid(np.dot(train_X,theta))
    grad = np.dot(train_X.T,(h_theta - train_y))
    
    theta = theta - (alpha)*grad
    J = (1/N)*(-np.dot(train_y.T,np.log(h_theta)) - np.dot((1-train_y.T),np.log(1-h_theta)))
    cost.append(J[0,0])
    #iterat.append(i)
    #print(theta,"\t",i)

y_pred = sigmoid(np.dot(train_X,theta))
```

```python
plt.plot(cost)
plt.ylabel("Loss")
plt.xlabel("epoch")
plt.title("Training Loss")
plt.show()
```

```python
y_pred = np.round(y_pred)

count=0
for i in range(len(train_y)):
    if train_y[i]!=y_pred[i]:
        count=count+1
print("Total ",count," out of",N,"are misclassified giving accuracy of",100*((N-count)/N),"%")
```

```python
# Confusion Matrix
from sklearn.metrics import confusion_matrix, precision_recall_fscore_support
cm = confusion_matrix(train_y, y_pred, labels = [0,1])
#print(cm)

f,ax = plt.subplots(figsize = (3, 3))
sns.heatmap(cm, annot = True, fmt = "d", ax = ax)
plt.show()


prf = precision_recall_fscore_support(train_y, y_pred, labels = [0,1])
print("Precision:",prf[0])
print("Recall:",prf[1])
print("F1 Score:",prf[2])
```

```python
# Prediction on test set
y_pred_test = sigmoid(np.dot(test_X,theta))

y_pred_test = np.round(y_pred_test)

count_t=0
for i in range(Nt):
    if test_y[i]!=y_pred_test[i]:
        count_t=count_t+1
print("Total ",count_t," out of",Nt,"are misclassified giving accuracy of",100*((Nt-count_t)/Nt),"%")
```

```python
# Confusion Matrix
from sklearn.metrics import confusion_matrix, precision_recall_fscore_support
cm = confusion_matrix(test_y, y_pred_test, labels = [0,1])
#print(cm)

f,ax = plt.subplots(figsize = (3, 3))
sns.heatmap(cm, annot = True, fmt = "d", ax = ax)
plt.show()

prf = precision_recall_fscore_support(test_y, y_pred_test, labels = [0,1])
print("Precision:",prf[0])
print("Recall:",prf[1])
print("F1 Score:",prf[2])
```
